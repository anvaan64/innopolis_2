```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array(list(range(20)))
x
```

```python
y = x**2
```

```python
plt.xlabel('X')
plt.ylabel('Squares')
plt.title('Title')
line = plt.plot(x,y, linewidth = 5, color='c', marker='^', alpha=0.4)
plt.savefig('123.png')
```

```python
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
```

```python
sns.distplot(data['total_bill'])
```

```python
# sns.pairplot(data, kind='kde')
sns.pairplot(data, hue='smoker')
```

```python
sns.lmplot(x = 'total_bill', y =  'size', data = data)
```

```python
sns.jointplot(x = 'total_bill', y =  'tip', data = data)
```

```python
sns.boxplot(x='day', y='total_bill', data=data)
```

```python
sns.boxplot(x='day', y='tip', data=data)
```
